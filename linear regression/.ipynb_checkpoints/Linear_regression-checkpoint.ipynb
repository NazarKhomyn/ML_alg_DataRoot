## Linear regression

```python
import numpy as np

%matplotlib inline
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import stochastic_gradient_descent as sgd
import batch_gradient_descent as bgd
```

```python
# Extract data
points = np.genfromtxt('data.csv',delimiter = ',')
n = points.shape[0]

X0 = np.ones((n, 1))
X1 = points[:,0].reshape(n,1)
X = np.hstack((X0, X1))

# Points y-coordinate
y = points[:,1].reshape(n,1)

# The Learning Rate alpha.
alpha = 0.001

#stochastic_gradient_descent
optimal = sgd.stochastic_gradient_descent(X, y, alpha,num_iter=100000)

print('optimal:', optimal)
print('error function:', sgd.error_function(optimal, X, y)[0,0])
```

```python
a, b = optimal[0],optimal[1]
```
